```python
"""
1. Building the k-means clustering algorithm:
a. Start off by extracting the ‘mpg’, ‘disp’ & ‘hp’ columns from the ‘mtcars’ data.frame.
Store the result in ‘car_features’
b. Build the kmeans algorithm on top of ‘car_features’. Here, the number of clusters
should be 3
c. Bind the clustering vector to ‘car_features’.
d. Extract observations belonging to individual clusters
"""
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("C:\\Users\\black\\Desktop\\PyforDS\\datasets\\mtcars.csv")
```

```python
data.head()
```

```python
car_features = data[["mpg","disp","hp"]]
```

```python
car_features.head()
```

```python
#Taking 3 clusters
km1=KMeans(n_clusters=3)

#Fitting the input data
km1.fit(car_features)

#predicting the labels of the input data
y=km1.predict(car_features)

#adding the labels to a column named label
data["label"] = y

#The new dataframe with the clustering done
data.head()
```

```python
# Observations belonging to cluster - 1
data[km1.labels_==0].head()
```

```python
# Observations belonging to cluster - 2
data[km1.labels_==1].head()
```

```python
# Observations belonging to cluster - 3
data[km1.labels_==2].head()
```

```python
"""
2. On the same ‘car_features’ dataset build a k-means algorithm, where the number of clusters is 5
a. Bind the clustering vector to ‘car_features’
b. Extract observations belonging to individual clusters 
"""
```

```python
#Taking 5 clusters
km2=KMeans(n_clusters=5)

#Fitting the input data
km2.fit(car_features)

#predicting the labels of the input data
y=km2.predict(car_features)

#adding the labels to a column named label
data["label"] = y

#The new dataframe with the clustering done
data.head()
```
